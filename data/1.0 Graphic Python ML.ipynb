```python
#Tuturial by Joao Victor.
import matplotlib.pyplot as plt
import numpy as np
# the library matplotlibgo make the grafics and numpy go run the numbers. 
# To begin with Machine Learn you must first know about python programming and test graphics.

#here we can begin with 'plot()'. 'plot()' will make the graphic in itself.
#the '[]' into that 'plot()'defined the value that the graph line will follow. in the sample i put 10 how the first value in the graph.
# if you put 20 or 30... in the value the graph will automatically extend in the size necessary
plt.plot( [10,5,3,4,6,8] )

#the 'title()' will show above the graph a title.
plt.title("First graph sample")

#the 'show()' will show the graphic.
plt.show()
```

```python
#Tuturial by Joao Victor.
#import matplotlib.pyplot as plt
#import numpy as np

#The second sample is a little more complex.
#here is defined the variables 'data1' & 'data2' with the values.
data1 = [10,5,2,4,6,8]
data2 = [ 1,2,4,8,7,4]

#here the 'x' will distribute the width of the values. for example the value 10 from 'data1' will start at 0, and the 8 will end at 50.
x = 10*np.array(range(len(data1)))

#here the graph is built with some instructions.
#the'x' is the width of values. the 'data1' is the values in itself.
plt.plot( x, data1, 'go') #'go' is easy to explain the 'g' = green and 'o' = format in circle.
plt.plot( x, data1, 'k:', color='orange') #'k:' is the line format,in this case the line will be dotted cause it has':'. and you define the collor.


plt.plot( x, data2, 'r^') #'r^' 'r' = red and '^' = format in triangle.
plt.plot( x, data2, 'k--', color='blue')#'k' is the line format,in this case the line will be dashed cause it has '--'. and you define the collor.


#the 'axis()' will convenience method to get or set some axis properties.
plt.axis([-10, 60, 0, 11])#'-10,60' define the horizontal limit, and '0,11' define the vertical limit.
plt.title("The second sample")

plt.grid(True)#the grid define if you want a checkered grid behind the graph.

#'x' represent horizontal and 'label' will give a name for that x axis. 
plt.xlabel("horizontal axis | x")

#'y' represent vertical and 'label' will give a name for that y axis. 
plt.ylabel("vertical axis | y")
plt.show()
```
