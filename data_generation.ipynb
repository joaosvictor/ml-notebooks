```python
from sklearn.datasets.samples_generator import make_regression
# X is the sample resource, y is the sample result, coef is the regression coefficient, a total of 1000 samples, 1 resource per sample
x, y, coef =make_regression(n_samples=1000, n_features=1,noise=10, coef=True)

# plotting
plt.scatter(x, y,  color='orange')
plt.plot(x, x*coef, color='blue'
         ,linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets.samples_generator import make_classification
# X1 is the sample resource, Y1 is the output of the sample category, a total of 400 samples, each sample has 2 resources, the output has 3 categories, without redundant resources and each category has a cluster
X1, Y1 = make_classification(n_samples=400, n_features=2, n_redundant=0,
                             n_clusters_per_class=1, n_classes=3)
plt.scatter(X1[:, 0], X1[:, 1], marker='.', c=Y1)
plt.show()
```

```python
from sklearn.datasets.samples_generator import make_blobs
# X is the sample feature, Y is the sample cluster category, a total of 1000 samples, each sample has 2 features, a total of 3 clusters, the cluster center is at [-1,-1], [1,1], [2,2 ], the cluster variances are [0.4, 0.5, 0.2]
X, y = make_blobs(n_samples=1000, n_features=2, centers=[[-1,-1], [1,1], [2,2]], cluster_std=[0.4, 0.5, 0.2])
plt.scatter(X[:, 0], X[:, 1], marker='.', c=y)
plt.show()
```

```python
from sklearn.datasets import make_gaussian_quantiles
#Generate a 2-dimensional normal distribution, the generated data is divided into 3 groups according to the quantile, 1000 samples, 2 sample feature mean values are 1 and 2, the covariance coefficient is 2
X1, Y1 = make_gaussian_quantiles(n_samples=1000, n_features=2, n_classes=3, mean=[1,2],cov=2)
plt.scatter(X1[:, 0], X1[:, 1], marker='.', c=Y1)
```
