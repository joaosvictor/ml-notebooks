```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import joblib
```

```python
data = pd.read_csv('school.csv')
data
```

```python
import seaborn as sb
sb.pairplot(data, hue='year')
```

```python
us=data[data.country == 'United State']
br=data[data.country == 'Brazil']
#plt.plot(us.year, us.number / us.number.iloc[0] * 1,'k:',color="navy")
#plt.plot(us.year, us.number, 'go')
plt.plot(us.year, us.number ,'k-',color="green")
#plt.plot(br.year, br.number, 'yo')
plt.plot(br.year, br.number,'k-', c='darkorange')
plt.legend(['United States.','Brazil.'])
plt.title(  '  Statistics of teenagers who dropout of high school')
plt.xlabel("Years.")
plt.ylabel("Nº in millions.")
plt.grid(False)
plt.show()
```
